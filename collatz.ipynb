# Conjectura de Collatz

A **Conjectura de Collatz**, tamb√©m conhecida como o problema 3x+1, √© uma das quest√µes matem√°ticas mais intrigantes e simples de se enunciar, mas ainda n√£o resolvidas. Proposta por Lothar Collatz em 1937, ela sugere que, dado um n√∫mero natural ùëõ, ao aplicar repetidamente as seguintes regras, sempre se chegar√° ao n√∫mero 1: 
* Se ùëõ for par, divida-o por 2; 
* Se ùëõ for √≠mpar, multiplique-o por 3 e some 1.

Embora extensivamente testada para n√∫meros muito grandes, ainda n√£o h√° uma demonstra√ß√£o matem√°tica que garanta que a conjectura seja verdadeira para todos os n√∫meros naturais. Essa simplicidade aparente combinada com a dificuldade de prova torna a conjectura um mist√©rio fascinante na teoria dos n√∫meros.

## Demonstrando

```python
# Lista para armazenar todas as sequ√™ncias
all_sequences = []

for c in range(1, 11):
    num = c
    sequence = [num]  # Inicia a sequ√™ncia com o n√∫mero inicial
    
    print(num, end=' -> ')

    while True:
        if num == 1:  # Verifica se o n√∫mero √© igual a 1
            print(num)
            break
        
        elif num % 2 == 0:  # Verifica se o n√∫mero √© par
            num = num // 2  # Usa divis√£o inteira para garantir que o resultado seja um inteiro

        elif num % 2 != 0:  # Verifica se o n√∫mero √© √≠mpar
            num = num * 3 + 1
        sequence.append(num)  # Adiciona o n√∫mero na sequ√™ncia
        
        if num != 1:
            print(num, end=' -> ')
    all_sequences.append(sequence)  # Salva a sequ√™ncia na lista
```

# An√°lise Estat√≠stica

```python
import numpy as np

# Exibe todas as sequ√™ncias salvas
for i, seq in enumerate(all_sequences, start=1):
    print(f"Sequ√™ncia {i}:")
    print(seq)
    
    # Calculando as m√©tricas estat√≠sticas
    mean = np.mean(seq)
    median = np.median(seq)
    std_dev = np.std(seq)
    min_value = np.min(seq)
    max_value = np.max(seq)
    tamanho = len(seq)

    # Arredondando os resultados para 2 casas decimais
    mean = round(mean, 2)
    median = round(median, 2)
    std_dev = round(std_dev, 2)
    min_value = round(min_value, 2)
    max_value = round(max_value, 2)

    # Exibindo os resultados
    print(f"Converge ap√≥s {tamanho} valores")
    print(f"M√©dia: {mean}")
    print(f"Mediana: {median}")
    print(f"Desvio padr√£o: {std_dev}")
    print(f"Valor m√≠nimo: {min_value}")
    print(f"Valor m√°ximo: {max_value}")
    
    print('-'*100)
```

# Representa√ß√£o Gr√°fica

```python
import matplotlib.pyplot as plt

# Fun√ß√£o para calcular a sequ√™ncia de Collatz
def collatz_sequence(start):
    sequence = [start]
    while start != 1:
        if start % 2 == 0:
            start //= 2
        else:
            start = start * 3 + 1
        sequence.append(start)
    return sequence

# Gerar e plotar as sequ√™ncias de 1 a 10
plt.figure(figsize=(10, 6))

for c in range(1, 11):
    sequence = collatz_sequence(c)
    plt.plot(range(len(sequence)), sequence, label=f"In√≠cio: {c}")

# Configura√ß√µes do gr√°fico
plt.title("Trajet√≥rias da Conjectura de Collatz (1 a 10)")
plt.xlabel("Passos")
plt.ylabel("Valor")
plt.legend()
plt.grid()

# Mostrar o gr√°fico
plt.show()

```
